### Preparing data for modelling

```python
revenue_contributed = df.groupby('Customer ID')['Sales'].sum()
revenue_contributed = revenue_contributed.reset_index()
revenue_contributed.head()
```

```python
frequency = df.groupby('Customer ID')['Order ID'].count()
frequency = frequency.reset_index()
frequency.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format= ' %d-%m-%Y')
df['duration'] = df['Order Date'].max() - df['Order Date']
df['duration']
```

```python
recency = df.groupby('Customer ID')['duration'].min()
recency = recency.reset_index()
recency.head()
```

```python
rf = pd.merge(revenue_contributed, frequency, on='Customer ID', how='inner')
rfm = pd.merge(rf, recency, on='Customer ID', how='inner')
rfm.columns = ['Customer ID', 'revenue', 'frequency', 'recency']
rfm['recency'] = rfm['recency'].dt.days
rfm.head()
```

### Outlier treatment

```python
plt.boxplot(rfm['revenue'])
plt.show()
plt.boxplot(rfm['frequency'])
plt.show()
plt.boxplot(rfm['recency'])
plt.show()
```

```python
q1_rev = rfm['revenue'].quantile(0.05)
q3_rev = rfm['revenue'].quantile(0.95)
IQR = q3_rev - q1_rev
rfm = rfm[(rfm['revenue'] >= q1_rev - 1.5*IQR) & (rfm['revenue'] <= q3_rev + 1.5*IQR)]

q1_rev = rfm['frequency'].quantile(0.05)
q3_rev = rfm['frequency'].quantile(0.95)
IQR = q3_rev - q1_rev
rfm = rfm[(rfm['frequency'] >= q1_rev - 1.5*IQR) & (rfm['frequency'] <= q3_rev + 1.5*IQR)]

q1_rev = rfm['recency'].quantile(0.05)
q3_rev = rfm['recency'].quantile(0.95)
IQR = q3_rev - q1_rev
rfm = rfm[(rfm['recency'] >= q1_rev - 1.5*IQR) & (rfm['recency'] <= q3_rev + 1.5*IQR)]
```

```python
### Scaling the features
scale = StandardScaler()
cols = ['revenue', 'frequency', 'recency']
rfm_scaler = scale.fit_transform(rfm[cols])
```

```python
rfm_scaled = pd.DataFrame(rfm_scaler)
rfm_scaled.columns = ['revenue','frequency','recency']
rfm_scaled.head()
```

### Hopkins Statistic

```python
def hopkins(X):
    d = X.shape[1]
    n = len(X)
    m = int(0.1 * n)
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
    
    rand_X = sample(range(0,n,1),m)
    
    ujd = []
    wjd = []
    for j in range(0,m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1,-1),2,return_distance=True)
        wjd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1,-1),2,return_distance=True)
        wjd.append(w_dist[0][1])
        
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd,wjd)
        H = 0
    
    return H
```

```python
hopkins(rfm_scaled)
```

```python
### building Model(Hierarchial Cluster)
```

```python
# complete linkage
plt.figure(figsize=(20,12))

mergings = linkage(rfm_scaled, method = "complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
#cluster
cluster_labels = cut_tree(mergings, n_clusters=3).reshape(-1, )
cluster_labels
```

```python
# assign cluster labels
pd.set_option('display.max_rows', None)
rfm['cluster_labels'] = cluster_labels
rfm.head(90)
```
