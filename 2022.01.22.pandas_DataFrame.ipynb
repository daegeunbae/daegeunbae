```python
# Q) 기존의 DF에서, 날짜별로 (lambda apply사용)
#                   최고가격과 최저가격의 차이를 계산해서
#                   그 값을 새로운 "Gap"컬럼에 만들어보세요~
# Q) GapFor컬럼에는 동일한 기능인데, for문을 통해서 작업!!
```

```python
stock_price_DF["gap"] = stock_price_DF.apply(lambda x : x.max() - x.min(), axis=1)
stock_price_DF
```

```python
stock_price_DF["gapfor"] = 0
stock_price_DF
```

```python
# 기준 : 정수 인덱스1!!!!!
for i in range(0, len(stock_price_DF)):
    가로 i번째 줄에서 앞의 4개의 값들 가져오기!!! : iloc
    4개의 값 중에서 max
    4개의 값 중에서 min
    이 값의 차이 : temp
    업데이트 할 위치 : 가로줄 i, 세로줄 gapfor 컬럼줄 -> temp
    
```

```python
stock_price_DF
```

```python
nrows = stock_price_DF.shape[0]
for i in range(nrows): # 0~4까지 i번째 가로줄 접근 인덱스
    data = stock_price_DF.iloc[i, :4]
    temp = data.max()-data.min()
    stock_price_DF.iat[i, 5] = temp
stock_price_DF
```

```python

cols = ["시가","종가","고가","저가"]
cols_idx = [ stock_price_DF.columns.get_loc(col) for col in cols]
cols_idx
```

```python
# 수정) 
stock_price_DF["gapfor"] = 0
nrows = stock_price_DF.shape[0]
idx = stock_price_DF.columns.get_loc("gapfor")

cols = ["시가","종가","고가","저가"]
cols_idx = [ stock_price_DF.columns.get_loc(col) for col in cols]

for i in range(nrows): # 0~4까지 i번째 가로줄 접근 인덱스
    data = stock_price_DF.iloc[i, cols_idx]
    temp = data.max()-data.min()
    stock_price_DF.iat[i, idx] = temp
stock_price_DF
```
