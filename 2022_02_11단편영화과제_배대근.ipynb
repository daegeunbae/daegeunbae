```python
## kobis api : json #220102 movieTypeCd
#2021년 조회시작으로 해서 : prdtStartYear=2021
#(아마 대충 5XX개 정도)
#-> step1) 100개에 대해서만: itemPerPage =100
#영화코드, 영화이름, 영문이름, 장르, 상영시간(필수), 국가, 감독이름, 배우도...
#-> step2) 전체 5XX개 영화에 대해서 step1의 정보를 수집~~

#필요한 패키지들
import pandas as pd
import urllib.request # urllib, requests
import json
```

```python
movie_df = pd.DataFrame()
movie_df_temp = pd.DataFrame()
movie_df_temp = movie_df_temp.append({"movieCd":"",
                                      "movieNm":"",
                                      "movieNmEn":"",
                                      "genreAlt":"", 
                                    #   "showTm":"", 
                                      "repNationNm":"", 
                                      "directors":""
                                    #   "actors":""
                                      },
                                 ignore_index=True)

for i in range(7):
    url_p1 = "http://kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key="
    key = "16017155671c20ec5a59a6a52b651447"
    url_p2 = "&itemPerPage=100" #100개씩 보겠다
    movie_code = "&prdtStartYear=2021" #요청 인터페이스에서 볼 코드값
    movieTypeCd = "&movieTypeCd=220102" ##단편영화찾겠다 사이트 열어서 영화유형코드 2201 조회
    curpage = "&curPage={}".format(i)
    url = url_p1 + key + url_p2 + movie_code + movieTypeCd + curpage
    temp_page = urllib.request.urlopen(url)
    temp_data = json.loads(temp_page.read().decode("utf-8"))
    # 내가 받은 영화의 수를 기반으로 해야하니
    num = len(temp_data["movieListResult"]["movieList"])
# 저는 i번째 영화에 대해서 정보를 추출..(정수)
    for j in range(0, num):
    # i번째 영화에 대해서
        i_movie = temp_data["movieListResult"]["movieList"][j]
    # print(i_movie)
    #       필요한 정보(코드,이름, 개봉일..)추출
        movie_df_temp.iat[0,0] = i_movie["movieCd"]
        movie_df_temp.iat[0,1] = i_movie["movieNm"]
        movie_df_temp.iat[0,2] = i_movie["movieNmEn"]
        movie_df_temp.iat[0,3] = i_movie["genreAlt"]
        movie_df_temp.iat[0,4] = i_movie["repNationNm"]
    # movie_df_temp.iat[0,5] = i_movie["directors"]

    # display(movie_df_temp)
    # movie_df_temp.iat[0,4] = i_movie["showTm"]
    # movie_df_temp.iat[0,7] = i_movie["actors"]

        if i_movie["directors"] !=[]: # 감독정보 있을 때.
            movie_df_temp.iat[0,5] = i_movie["directors"][0]["peopleNm"]

    #       -> 단, 정보가 없을수도;;;예외처리!!!
    #       temp에 기록!!!(가로:i번째, 세로:그 정보 컬럼 정수인덱스)

    # temp것을 movie_df에 추가~~~
        movie_df = movie_df.append(movie_df_temp,ignore_index=True)

    # temp싹 방정리~~ -> 다음 영화 정보르..
        for j in range(len(movie_df_temp.columns)):
            movie_df_temp.iat[0, j] = ""
```
