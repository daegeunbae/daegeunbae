```python
import requests
import numpy as np
import pandas as pd

# row 출력제한 해제
pd.set_option('display.max_row', 500)

# column 출력제한 해제
pd.set_option('display.max_columns', 100)
import datetime
 
import time
import re
import selenium
from selenium import webdriver as wd 
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
# chromedriver 있는 위치 지정하기. 
from selenium.common.exceptions import NoSuchElementException
dt_now = datetime.datetime.now()
```

```python
a=[];b=[];c=[];d=[];e=[]
div = wd.Chrome('C:/Users/gmddj/Downloads/chromedriver_win32 (1)/chromedriver.exe')

for j in range(1,10):
    div.get('http://www.e-newsp.com/mobile/section_list_all.html?sec_no=74&page='+str(j))
    for i in range(1,14):
        path1 = div.find_element(By.CSS_SELECTOR,'#container > div.con_primary > div > div.m_ara_001 > ul.list_thumb > li:nth-child('+str(i)+') > a > h3').text
        path2 = div.find_element(By.CSS_SELECTOR,'#container > div.con_primary > div > div.m_ara_001 > ul.list_thumb > li:nth-child('+str(i)+') > a > ul > li:nth-child(1)').text
        path3 = div.find_element(By.CSS_SELECTOR,'#container > div.con_primary > div > div.m_ara_001 > ul.list_thumb > li:nth-child('+str(i)+') > a > span > img').get_attribute('src')
        path4 = div.find_element(By.CSS_SELECTOR,'#container > div.con_primary > div > div.m_ara_001 > ul.list_thumb > li:nth-child('+str(i)+') > a > ul > li:nth-child(2)').text
        div.find_element(By.CSS_SELECTOR,'#container > div.con_primary > div > div.m_ara_001 > ul.list_thumb > li:nth-child('+str(i)+') > a > h3').click()
        path5 = div.current_url
        div.back()
        a.append(path1)
        b.append(path2)
        c.append(path3)
        d.append(path4)
        e.append(path5)
    time.sleep(1)

    
    
ex=pd.DataFrame(zip(a,b,c,d,e), columns = ['Title','Date','Image','Reporter','URL'])
ex
```
